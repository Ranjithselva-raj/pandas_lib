```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
titanic
```

```python
#drop the null values in the dataset columns = axis 1 and for rows = axis 0
titanic.dropna(axis=1)
```

```python
# To check how many nulll values in each columns need to use the below code

titanic.isnull().sum()
```

```python
# when like to drop th columns which have more than certain percentage of null values : use Thresh 

titanic.dropna(axis=1, thresh= 800)
```

```python
# if we want to drop the null values if the null values present in all the rows use how = all 

titanic.dropna(axis=1, how='all')
```

```python
#to fill the null values with 0
titanic.fillna(value=0)
```

```python
pd.DataFrame({
    'Col1': [1,2,3,4,5],
    'Col2': [45,67,34,12,34],
    'Col3': "I like to learn coding"
})
```

```python
df=pd.DataFrame({
    'Col1': [1,2,3,4,5],
    'Col2': [45,67,34,12,34],
    'Col3': "I like to learn coding".split()
})
```

```python
df
```

```python
df['Col4']=df['Col1']**2
```

```python
df
```

```python
df['Col5']=df['Col3'].str[0]
```

```python
df
```

```python
def sqrt(x):
    return x**2
```

```python
df['col4_4']=df['Col1'].apply(sqrt)
```

```python
df
```

```python
def filt(x):
    if x == 'I':
        return 'I'
    else:
        return 'Z' 
```

```python
df['Col6']=df['Col5'].apply(filt) 
```

```python
df
```

```python
df['Col6_6']=df['Col5'].apply(lambda x: 'I' if x == 'I' else 'Z')
```

```python
df
```

```python
def coun(x):
    return len(x)
```

```python
#add a new column that give the lenght of the string present in Col3
df['Col7']=df['Col3'].apply(coun)
```

```python
df
```

```python
#add a new column that give the lenght of the string present in Col3 using lambda function

df['Col7_7']=df['Col3'].apply(lambda x: len(x))
```
