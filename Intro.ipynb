```python
key_column=[]
for i in data1[0]['user'].keys():
    key_column.append(i)

key_column
```

```python
login=[]
id=[]
node_id=[]
avatar_url=[]
gravatar_id=[]
url=[]
html_url=[]
followers_url=[]
following_url=[]
gists_url=[]
starred_url=[]
subscriptions_url=[]
organizations_url=[]
repos_url=[]
events_url=[]
received_events_url=[]
type=[]
site_admin=[]

try:

    for i in range(len(data1)):
        for j in key_column:
            if j=='login':
                login.append(data1[i]['user'][j])     
            elif j=='id':
                id.append(data1[i]['user'][j])
            elif j=='node_id':
                node_id.append(data1[i]['user'][j])
            elif j=='avatar_url':
                avatar_url.append(data1[i]['user'][j])
            elif j=='gravatar_id':
                gravatar_id.append(data1[i]['user'][j])
            elif j=='gravatar_url':
                url.append(data1[i]['user'][j])
            elif j=='html_url':
                html_url.append(data1[i]['user'][j])
            elif j=='followers_url':
                followers_url.append(data1[i]['user'][j])
            elif j=='following_url':
                following_url.append(data1[i]['user'][j])
            elif j=='gists_url':
                gists_url.append(data1[i]['user'][j])
            elif j=='starred_url':
                starred_url.append(data1[i]['user'][j])
            elif j=='subscriptions_url':
                subscriptions_url.append(data1[i]['user'][j])
            elif j=='organizations_url':
                organizations_url.append(data1[i]['user'][j])
            elif j=='repos_url':
                repos_url.append(data1[i]['user'][j])
            elif j=='events_url':
                events_url.append(data1[i]['user'][j])
            elif j=='received_events_url':
                received_events_url.append(data1[i]['user'][j])
            elif j=='type':
                type.append(data1[i]['user'][j])
            elif j=='site_admin':
                site_admin.append(data1[i]['user'][j])
    else:
        print('Code Completed')
except Exception as e:
    print(e)

    
data_dict={
    'login':login,
    'id':id,
    'node_id':node_id,
    'avatar_url':avatar_url,
    'gravatar_id':gravatar_id,
    'html_url':html_url,
    'followers_url':followers_url,
    'following_url':following_url,
    'gists_url':gists_url,
    'starred_url':starred_url,
    'subscriptions_url':subscriptions_url,
    'organizations_url':organizations_url,
    'repos_url':repos_url,
    'events_url':events_url,
    'received_events_url':received_events_url,
    'type':type,
    'site_admin':site_admin
    

}

df=pd.DataFrame(data_dict)
df
```
