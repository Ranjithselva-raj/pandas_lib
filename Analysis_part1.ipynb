```python
#1. Find out how many male and female passenger was onborded

passenger_count = git_titanic['Sex'].groupby(git_titanic['Sex']).count().rename('Onboarded').reset_index()

passenger_count

```

```python
#2.how many survived we have

Survived = git_titanic.groupby(['Sex'])['Survived'].sum().rename('Survived').reset_index()

Survived
```

```python
# How many casualities we have


Casualities =git_titanic[git_titanic['Survived']==0].groupby('Sex').size().rename('Casualities').reset_index()

Casualities
```

```python
summary=pd.concat([passenger_count,Survived],axis=1)

summary
```

```python
merged=passenger_count.merge(Survived,on='Sex')

summ=merged.merge(Casualities,on='Sex')

summ
```

```python
#how many passenger do we have in first,second and third class

Passenger_count_class=git_titanic.groupby(git_titanic['Pclass'])['PassengerId'].count().rename('Passenger_count').reset_index()

Passenger_count_class
```

```python
#8. how many person do we have below age of 25

git_titanic[git_titanic['Age']<25]['PassengerId'].count()
```

```python
#9.how many person died whose age was less than 40

filtered_data = git_titanic[(git_titanic['Age'] < 40) & (git_titanic['Survived'] == 0)]

# Calculate the count of 'PassengerId' for the filtered data
count_result = filtered_data['PassengerId'].count()

print('person died whose age was less than 40: ',count_result)

```
